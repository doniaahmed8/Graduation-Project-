Check unique values 

```python

for column in df.columns:
    unique_values = df[column].unique()
    print(f"Unique values in '{column}': {unique_values}")
```

Check for low variance 

```python

feature_columns = [col for col in df.columns if col not in ['SHA256', 'Type']]

ones_count = df[feature_columns].sum()

print("Total count of '1's for each feature:")
print(ones_count)
```

```python
columns_to_drop = ones_count[ones_count < 500].index

df = df.drop(columns=columns_to_drop)

print(f"Remaining columns after dropping those with fewer than 500 '1's: {df.columns}")
```

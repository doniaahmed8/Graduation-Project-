Check Correlation

```python

corr_matrix = New_df.corr()

highly_correlated_mask = (corr_matrix > 0.8) & (corr_matrix < 1.0)

highly_correlated_pairs = []
for i in range(len(highly_correlated_mask.columns)):
    for j in range(i):
        if highly_correlated_mask.iloc[i, j]:
            highly_correlated_pairs.append((highly_correlated_mask.columns[i], highly_correlated_mask.columns[j]))

features_to_remove = set()
for feature1, feature2 in highly_correlated_pairs:
    if feature1 not in features_to_remove:

        features_to_remove.add(feature2)
features_to_remove

```

```python
df_filtered = New_df.drop(columns=features_to_remove)
```
